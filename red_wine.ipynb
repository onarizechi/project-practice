```python
plt.figure(figsize=(20, 25))
p = 1

# Loop through each column in the DataFrame
for i in df.columns:
    if p <= 11 and i != 'fixed acidity':  # Skip 'fixed acidity' itself
        plt.subplot(5, 4, p)
        sns.regplot(x='fixed acidity', y=i, data=df, color='r')
        plt.xlabel("fixed acidity")
        plt.ylabel(i)
        p += 1  # Increment p only if a subplot is created

plt.tight_layout()  # Adjust subplot layout
plt.show()
```

```python
sns.pairplot(data = df,palette = "Dark2")
plt.show()
```

```python
# checking for outliers by plotting boxplot
plt.figure(figsize = (20,25),facecolor ='white')
plotnumber = 1
for i in df:
    if plotnumber<=11:
        plt.subplot(5,4,plotnumber)
        sns.boxplot(df[i],palette = "Set2_r")
        plt.xlabel(i)
        
    plotnumber+=1
plt.tight_layout()
plt.show()
```

We have found outliers in all the columns and will be removing using them using zscore and IQR techniques

```python
from scipy.stats import zscore
out_features = df [['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 
       'pH', 'sulphates', 'alcohol', 'quality']]
z = np.abs(zscore(out_features))
z
```

```python
df1 = df[(z<3).all(axis = 1)]
df1.shape
```

```python
#lets visualize how the data has been distributed in children column after removing skewness
plt.figure(figsize = (20,25),facecolor ='green')
plotnumber = 1
for i in df:
    if plotnumber<=11:
        plt.subplot(5,4,plotnumber)
        sns.distplot(df[i],color = "b")
        plt.xlabel(i, fontsize = 20)
        
    plotnumber+=1
plt.tight_layout()
plt.show()
```

```python
df.skew()
```

```python
#removing skewness using cuberoot method
df['residual sugar'] = np.cbrt(df['residual sugar'])
df['chlorides'] = np.cbrt(df['chlorides'])
df['free sulfur dioxide'] = np.cbrt(df['free sulfur dioxide'])
df['total sulfur dioxide'] = np.cbrt(df['total sulfur dioxide'])
df['sulphates'] = np.cbrt(df['sulphates'])
```
